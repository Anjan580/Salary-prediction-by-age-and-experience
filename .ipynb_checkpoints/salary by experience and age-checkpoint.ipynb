```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data loading 

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

## EDA

### Scatter plot using matplotlib ==> salary VS years of experience

```python
plt.scatter(df["YearsExperience"], df["Salary"], color ="red", edgecolor = "black")
plt.xlabel("years of experience", fontweight = "bold")
plt.ylabel("Salary", fontweight = "bold")
plt.show()
```

### Scatter plot using Seaborn ==> salary VS years of experience

```python
sns.scatterplot(x= "YearsExperience", y = "Salary", data = df, color = "blue", edgecolor = "black")
plt.xlabel("Years of Experience")
plt.show()
```

```python
corr = df.corr(numeric_only = True)
sns.heatmap(corr, annot = True, cmap = "coolwarm")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[["YearsExperience"]]
y = df["Salary"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
print(model.predict([[3]]))
```

```python
a = int(input("Enter the years of expereience"))
print(model.predict([[a]]))
```

```python
y_pred = model.predict(xtest)
```

# evalution 

### R2_score

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(ytest, y_pred)
```

```python
print(r2)
```

### MAE (Mean absolute error)

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(ytest, y_pred)
```

```python
print(mae)
```

### MSE (mean square error)

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(ytest, y_pred)
```

```python
print(mse)
```

### RMSE (Root mean Squared Error)

```python
rmse = np.sqrt(mse)
print(rmse)
```

# Multiple Regression ==> salary by age and experience 

```python
a = df[["YearsExperience","Age"]]
b = df["Salary"]
```

```python
atrain, atest, btrain , btest = train_test_split(a,b, test_size = 0.2, random_state = 10)
```

```python
multiplemodel = LinearRegression()
```

```python
multiplemodel.fit(atrain,btrain)
```

```python
prediction = multiplemodel.predict(atest)
```

## Evaluation

### r2_score

```python
r_2 = r2_score(btest, prediction)
print(r_2)
```

### MAE

```python
mae_= mean_absolute_error(btest,prediction)
print(mae_)
```

### MSE

```python
mse_ = mean_squared_error(btest, prediction)
print(mse_)
```

### RMSE

```python
rmse_ = np.sqrt(mse)
print(rmse_)
```

#### now lets predict 

```python
e = float(input("Enter the Years of Experience:"))
ag = int(input("Enter the age:"))
result = multiplemodel.predict([[e,ag]])
print(result.astype(float))
```
